```python
import pandas as pd
import numpy as np
```

## Lineage List (cov-lineages.org)

Data Source: https://cov-lineages.org/lineage_list.html

https://github.com/cov-lineages/lineages-website/blob/master/_data/lineage_data.full.json

```python
file_Path = "../Data/CovLineage/lineage_data.full.json"
```

```python
df = pd.read_json (file_Path).T
df
```

```python
df.info()
```

```python
df.columns
```

```python
data = df[['Lineage', 'Countries', 'Earliest date',
       'Latest date', 'Description']]
```

```python
data
```

```python
#Replace Blank/Empty String with NaN
data = data.replace('',np.nan,regex=True)
data
```

```python
data.isnull().mean()
```

```python
data = data.dropna()
data.isnull().mean()
```

```python
data
```

```python
most_common_lst = data.Countries.str.split(",").to_list()
len(most_common_lst)
```

```python
common_country_1 = [x[0].rsplit(' ', 1) for x in most_common_lst]
len(common_country_1)
```

```python
country_1 = [x[0] for x in common_country_1]
country_rate_1 = [float(x[1][:-1]) for x in common_country_1]
```

```python
len(country_1)
```

```python
# add it the info of most common country into dataframe
data['Most common country'] = country_1 
data['Most common country rate'] = country_rate_1
data
```

```python
data
```

```python
data = data[['Lineage', 'Most common country', 'Most common country rate', 'Earliest date', 'Latest date',
       ]]
```

```python
data
```

```python
output_f = "../Data/CovLineage/lineage_data.csv"
data.to_csv(output_f, index=False)
```
