```python
# select locations where have enough sequenceing data in NCBI
location_selected = ['Djibouti','Gambia','Egypt','Kenya','Seychelles','Nigeria','South Africa','Tunisia',
                     'Bahrain','Japan','Lebanon','Thailand','Vietnam','Iran','India','Bangladesh','Hong Kong',
                     'Denmark','France','Germany','Iceland','Slovakia','Switzerland','Italy','Estonia','Poland','United Kingdom',
                     'Mexico','United States','Canada',
                     'New Zealand',
                     'Brazil','Chile','Colombia','Paraguay']
len(location_selected)
```

```python
# check exists
[x in locations for x in location_selected]
```

```python
def get_csvBylocation(df, year, location_selected):
  for loc in location_selected:
    file_name = './'+ str(year) +'/'+ str(year) + '_' + loc + '.csv'
    data = df[df.location == loc]
    data.to_csv(file_name, index = True, encoding='utf-8')
```

```python
get_csvBylocation(df_2020, 2020, location_selected)
```

```python
df_2021 = df_owid.loc['2021-01-01':'2021-12-31']
df_2021.shape
```

```python
get_csvBylocation(df_2021, 2021, location_selected)
```

```python
df_2022 = df_owid.loc['2022-01-01':'2022-12-31']
df_2022.shape
```

```python
get_csvBylocation(df_2022, 2022, location_selected)
```
