### For Climate Data

```python
from pandas.core import frame
import glob
```

```python
all_files = glob.glob("../Data/Climate/nasa_data/*.csv")
li = []

for filename in all_files:
    climate = pd.read_csv(filename)
    cols = ['YEAR', 'MO', 'DY']
    climate['date'] = climate[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
    #print(climate.shape)
    climate['country'] = filename[26:-4]
    #print(climate.shape)
    li.append(climate)
    #print(len(li))
data2 = pd.concat(li, axis=0, ignore_index=True)
data2
```

```python
data2[['country', 'state']] = data2['country'].str.split('_', 1, expand=True)
```

```python
data2.replace('USA','United States', inplace=True)
data2.rename(columns={'ALLSKY_SFC_SW_DWN':'sky_shortwave_irradiance','WS2M':'wind_speed','PRECTOTCORR':'precipitation',
                     'RH2M': 'relative_humidity','QV2M': 'specific_humidity',
                     'T2M_RANGE':'temperature','WS50M_RANGE':'wind_speed_50meter_srange', 'WS10M_RANGE':'wind_speed_10meter_srange'},inplace=True)
data2
```

```python
data2.drop(columns='PS', inplace=True)
```

```python
cols = ['date', 'country']
data2['id'] = data2[cols].apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
```

```python
data2.shape
```

```python
data1.shape
```

data1.set_index('id', inplace=True)
data2.set_index('id', inplace=True)

```python
data = data2.merge(data1, left_on = "id", right_on = "id", how = "left")
data.shape
```

```python
data.columns
```

```python
data.drop(columns=['date_x','location'], inplace=True)
data.rename(columns={'date_y':'date'},inplace=True)
data.shape
```

```python
data.to_csv('./DataImportCSV/LocationDay.csv', index= False)
```
